```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from scipy.stats import chi2

# Example data setup (replace with your own data)
# Assuming inputs_train is your predictor variables and target_train is your target variable
# inputs_train should be a pandas DataFrame with your features

# Initialize Logistic Regression model
model = LogisticRegression(max_iter=1000)

# Fit the model using scikit-learn
model.fit(inputs_train, target_train)

# Extract coefficients and standard errors
coefficients = model.coef_[0]
n = len(target_train)  # number of observations
p = inputs_train.shape[1]  # number of features

# Predict using the model to get residuals
predictions = model.predict(inputs_train)
residuals = target_train - predictions

# Calculate Mean Squared Error (MSE)
MSE = np.sum(residuals**2) / (n - p - 1)

# Calculate variance of coefficients
variance = MSE * np.linalg.inv(np.dot(inputs_train.T, inputs_train)).diagonal()

# Calculate standard errors
std_errors = np.sqrt(variance)

# Calculate z-scores
z_scores = coefficients / std_errors

# Calculate two-tailed p-values
p_values = [2 * (1 - chi2.cdf(np.abs(z), 1)) for z in np.abs(z_scores)]

# Create summary table
summary_table = pd.DataFrame({
    'Feature Name': inputs_train.columns,
    'Coefficient': coefficients,
    'Standard Error': std_errors,
    'Z-Score': z_scores,
    'P-Value': p_values
})

# Print summary table
print("\nSummary Table with P-Values:")
print(summary_table)
```
